```python
import pandas as pd 
import matplotlib.pyplot as pypl
import numpy as np 
import seaborn as sd
```

```python
df = pd.read_excel('Total Servant Database.xlsx',index=False)
df.sample()
```

```python
df = df.drop(['ID','Servant Name','Japanese Name','AKA','NP Rank','NP Classification','NP Damage Type','NP Hit-Count','Traits'],axis=1)
```

```python
df['Class'].unique()
```

```python
df = df[df['Class'] != 'Beast II']
```

```python
sd.color_palette('Paired')
```

```python
pypl.figure(figsize=(18,9))
sd.color_palette('Paired')
sd.countplot('Class',data=df,palette='Paired')
```

```python
pypl.figure(figsize=(18,9))
sd.countplot('Rarity',data=df,hue='Class',palette='Paired')
```

```python
df.columns
```

```python
fig,(ax1,ax2) = pypl.subplots(2,figsize=(18,15))

ax1.hist(df['HP at level 1'],bins=20)
ax1.set(xlabel='HP at level 1',ylabel='Count')

ax2.hist(df['HP at max Servant level'],bins=20)
ax2.set(xlabel='HP at max Servant level',ylabel='Count')

pypl.show()
```

```python
fig, (ax1,ax2) = pypl.subplots(2,figsize=(18,15))

ax1.hist(df['ATK at level 1'],bins=30)
ax1.set(xlabel = 'ATK at level 1',ylabel='Count')

ax2.hist(df['ATK at max Servant level'],bins=30)
ax2.set(xlabel = 'ATK at max Servant level',ylabel='Count')

pypl.show()
```

```python
pypl.figure(figsize=(18,9))
sd.countplot('Voice Actor',data=df,order = pd.value_counts(df['Voice Actor']).iloc[:10].index)
```

```python
pypl.figure(figsize=(18,9))
sd.countplot('Illustrator',data=df,order = pd.value_counts(df['Illustrator']).iloc[:10].index)
```
