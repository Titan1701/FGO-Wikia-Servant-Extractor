```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as pypl
import seaborn as sd
```

```python
fgo_df = pd.read_csv('Total Servant Database.csv')
fgo_df.head()
```

```python
fgo_df = fgo_df[fgo_df['ID'] != 149]
```

```python
fgo_df.isnull().sum()
```

```python
fgo_df['Growth Curve'] = fgo_df['Growth Curve'].fillna('Reverse S')
```

```python
fgo_df.isnull().sum()
```

```python
fgo_df.columns
```

```python
new_col_names = {'ATK at level 1':'lvl 1 ATK',
                 'ATK at max Servant level' : 'Max lvl ATK',
                 'HP at level 1':'lvl 1 HP',
                 'HP at max Servant level' : 'Max lvl HP'}

fgo_df = fgo_df.rename(columns=new_col_names)
fgo_df.columns
```

```python
fgo_df['Growth Curve'].unique()
```

```python
fgo_df['Growth Curve'] = fgo_df['Growth Curve'].replace({'Semi-S':'Semi S','Reverse-S':'Reverse S','Semi-Reverse S':'Semi Reverse S'})
```

```python
fgo_df[['Star Generation', 'NP Charge ATK', 'NP Charge DEF', 'Death Rate']] = \
fgo_df[['Star Generation', 'NP Charge ATK', 'NP Charge DEF', 'Death Rate']].replace(r'%.*','',regex=True).apply(lambda x: x.str.strip())
```

```python
fgo_df[['Star Generation', 'NP Charge ATK', 'NP Charge DEF', 'Death Rate']] = \
fgo_df[['Star Generation', 'NP Charge ATK', 'NP Charge DEF', 'Death Rate']].astype(float)

fgo_df[['Star Generation', 'NP Charge ATK', 'NP Charge DEF', 'Death Rate']] = \
fgo_df[['Star Generation', 'NP Charge ATK', 'NP Charge DEF', 'Death Rate']].apply(lambda x : x/100)
```

```python
fgo_df[['Star Generation', 'NP Charge ATK', 'NP Charge DEF', 'Death Rate']]
```
